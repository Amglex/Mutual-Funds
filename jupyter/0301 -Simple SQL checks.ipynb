```python
import wrds
import feather
import pandas as pd
```

```python
# Connect to DB
db = wrds.Connection(wrds_username='amglex')
print('Successfully connected')
```

```python
# SQL Query
data_raw_df = db.raw_sql(
    '''
    SELECT holdings.crsp_portno, report_dt , percent_tna , crsp_company_key, fund_hdr.crsp_fundno , crsp_obj_cd 
    FROM holdings, fund_hdr , fund_style 
    WHERE  holdings.crsp_portno = fund_hdr.crsp_portno 
    AND fund_hdr.crsp_fundno = fund_style.crsp_fundno 
    AND report_dt = '2011-10-31'
    AND holdings.crsp_portno = 1028787
    '''
)
```

```python
print('\n')
print('Port_no:      {:^10s}'.format(str(data_raw_df.iloc[0,0])))
print('Report_dt:    {:^10s}'.format(str(data_raw_df.iloc[0,1])))
print('Obj_cd:       {:^10s}'.format(str(data_raw_df.iloc[0,5])))
print('\n')
print(data_raw_df.iloc[:,[2,3]])
```
